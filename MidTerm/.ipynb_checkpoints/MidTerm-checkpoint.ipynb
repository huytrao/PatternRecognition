```python
bands = {'alpha': (8, 13), 'delta': (0.5, 4), 'beta': (13, 30), 'gamma': (30, np.inf)}
```

```python
def get_data(filename):
    mat = scipy.io.loadmat(os.path.join(data_root, filename))
    data = mat['o']['data'][0, 0]
    FS = mat['o']['sampFreq'][0][0][0][0]

    states = {
     'focused': data[:FS * 10 * 60, :],
      'unfocused': data[FS * 10 * 60:FS * 20 * 60, :],
      'drowsy': data[FS * 30 * 60:, :],
    }
    return states
```

```python
def get_powers(channel, FS=128):
    # Loi b NaN hoc kim tra mng trng
    if channel.size == 0 or np.all(np.isnan(channel)):
        return {band: 0 for band in bands.keys()}  # Gi tr 0 nu khng c d liu hp l

    # Tin hnh tnh ton
    channel = channel - channel.mean()  # Trung bnh ha
    freq, psd = signal.periodogram(channel, fs=FS, nfft=256)

    powers = {}
    for band_name, band_limits in bands.items():
        low, high = band_limits
        band_psd = psd[(freq >= low) & (freq < high)]
        if band_psd.size > 0:
            powers[band_name] = band_psd.mean()
        else:
            powers[band_name] = 0  # Nu khng c gi tr, gn v 0
    return powers
```

```python
# ng dn n th mc cha cc tp .mat
data_root = 'eeg_data'

# Ly tt c cc tn tp trong th mc (gi s tp c nh dng "eeg_recordX.mat")
all_files = os.listdir(data_root)

# Lc ra cc tp c nh dng "eeg_recordX.mat"
subjects = [int(f.split('eeg_record')[1].split('.mat')[0]) for f in all_files if f.startswith('eeg_record')]

# In danh sch cc subject
print(subjects)
```

```python
from scipy import signal

rows_list = []
for subject_idx in range(1, 35):
    states = get_data(f"eeg_record{subject_idx}.mat")
    for ch_name, ch_idx in channel_map.items():
        for state, eeg in states.items():
            powers = get_powers(eeg[:, ch_idx])
            powers['state'] = state
            powers['channel'] = ch_name
            powers['subject'] = f"subject_{subject_idx}"
            rows_list.append(powers)
```

```python
df = pd.DataFrame.from_dict(rows_list)
```

```python
import pandas as pd

def remove_outliers_iqr(df, columns):
    """
    Loi b outliers cho cc ct ch nh trong DataFrame bng phng php IQR (Interquartile Range).
    
    Args:
    - df: DataFrame cha d liu
    - columns: Danh sch cc tn ct cn kim tra v loi b outliers
    
    Returns:
    - DataFrame  loi b outliers
    """
    for col in columns:
        Q1 = df[col].quantile(0.25)  # Tnh Q1
        Q3 = df[col].quantile(0.75)  # Tnh Q3
        IQR = Q3 - Q1  # Tnh khong IQR
        
        lower_bound = Q1 - 1.5 * IQR  # Gii hn di
        upper_bound = Q3 + 1.5 * IQR  # Gii hn trn
        
        # Lc b cc gi tr ngoi gii hn
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    
    return df

columns_to_check = ['alpha', 'delta', 'beta', 'gamma']
df_cleaned = remove_outliers_iqr(df, columns_to_check)

# In DataFrame  loi b outliers
df_cleaned.head()
```

```python

# Gi s df l DataFrame cha cc ct 'alpha', 'delta', 'beta', 'gamma', 'channel'
# Ly danh sch cc knh
channels = df_cleaned['channel'].unique()

# To figure
plt.figure(figsize=(15, 10))

# Duyt qua tng c trng
features = ['alpha', 'delta', 'beta', 'gamma']
for feature in features:
    feature_values = []
    
    # Tnh trung bnh gi tr ca c trng theo tng knh
    for ch in channels:
        mean_value = df_cleaned[df_cleaned['channel'] == ch][feature].mean()
        feature_values.append(mean_value)

    # V ng biu din gi tr trung bnh c trng ca tng knh
    plt.plot(channels, feature_values, label=f'{feature.title()} Power')

# Thm nhn, tiu , v ch thch
plt.xlabel('EEG Channels', fontsize=14)
plt.ylabel('Mean Power', fontsize=14)
plt.title('Mean Power of EEG Bands across Channels', fontsize=16)
plt.legend(loc='upper right', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)  # Xoay nhn knh nu cn
plt.tight_layout()

# Hin th  th
plt.show()
```
