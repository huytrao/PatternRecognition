```python
import time
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Hm  train v evaluate model
def train_and_evaluate_svm(X_train, X_test, y_train, y_test, kernel='linear', C=1.0):
    # Ghi li thi gian bt u
    start_time = time.time()
    
    # Khi to model
    svm = SVC(kernel=kernel, C=C)
    
    # Train model
    print(f"\nTraining SVM with {kernel} kernel...")
    svm.fit(X_train, y_train)
    train_time = time.time() - start_time
    
    # Thi gian prediction
    start_time = time.time()
    y_pred = svm.predict(X_test)
    predict_time = time.time() - start_time
    
    # Tnh accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # In kt qu
    print(f"\nKt qu cho SVM vi {kernel} kernel:")
    print(f"Thi gian training: {train_time:.2f} giy")
    print(f"Thi gian prediction: {predict_time:.2f} giy")
    print(f"Accuracy: {accuracy:.4f}")
    
    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, 
                              target_names=class_names))
    
    return svm, y_pred, train_time, predict_time, accuracy
```

```python
# Visualize confusion matrix
def plot_confusion_matrix(y_true, y_pred, title):
    plt.figure(figsize=(10, 8))
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names,
                yticklabels=class_names)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Train v evaluate Linear SVM
linear_svm, linear_pred, linear_train_time, linear_pred_time, linear_acc = \
    train_and_evaluate_svm(X_train, X_test, y_train, y_test, kernel='linear')
# Plot confusion matrices
plot_confusion_matrix(y_test, linear_pred, 'Confusion Matrix - Linear SVM')
```

```python
# Train v evaluate RBF SVM
rbf_svm, rbf_pred, rbf_train_time, rbf_pred_time, rbf_acc = \
    train_and_evaluate_svm(X_train, X_test, y_train, y_test, kernel='rbf')
plot_confusion_matrix(y_test, rbf_pred, 'Confusion Matrix - RBF SVM')
```

```python
# So snh thi gian v accuracy
models = ['Linear SVM', 'RBF SVM']
train_times = [linear_train_time, rbf_train_time]
pred_times = [linear_pred_time, rbf_pred_time]
accuracies = [linear_acc, rbf_acc]

# Plot comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Time comparison
x = range(len(models))
width = 0.35

ax1.bar(x, train_times, width, label='Training Time')
ax1.bar([i + width for i in x], pred_times, width, label='Prediction Time')
ax1.set_ylabel('Time (seconds)')
ax1.set_title('Computation Time Comparison')
ax1.set_xticks([i + width/2 for i in x])
ax1.set_xticklabels(models)
ax1.legend()

# Accuracy comparison
ax2.bar(models, accuracies)
ax2.set_ylabel('Accuracy')
ax2.set_title('Model Accuracy Comparison')
ax2.set_ylim([0, 1])

plt.tight_layout()
plt.show()

# In kt qu tng hp
print("\nKt qu tng hp:")
results_df = pd.DataFrame({
    'Model': models,
    'Training Time (s)': train_times,
    'Prediction Time (s)': pred_times,
    'Accuracy': accuracies
})
print(results_df.to_string(index=False))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import time

# 1. Thc hin PCA
def apply_pca(X_train, X_test, n_components):
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    
    # Tnh explained variance ratio
    explained_variance_ratio = pca.explained_variance_ratio_
    cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
    
    return X_train_pca, X_test_pca, explained_variance_ratio, cumulative_variance_ratio

# p dng PCA vi s components khc nhau
n_components_list = [50, 100, 200]
pca_results = {}

for n in n_components_list:
    X_train_pca, X_test_pca, var_ratio, cum_var_ratio = apply_pca(X_train, X_test, n)
    pca_results[n] = {
        'X_train': X_train_pca,
        'X_test': X_test_pca,
        'var_ratio': var_ratio,
        'cum_var_ratio': cum_var_ratio
    }

# 2. Thc hin LDA
def apply_lda(X_train, X_test, y_train, y_test, n_components):
    lda = LinearDiscriminantAnalysis(n_components=n_components)
    X_train_lda = lda.fit_transform(X_train, y_train)
    X_test_lda = lda.transform(X_test)
    return X_train_lda, X_test_lda

# LDA ch c th gim xung s classes - 1
X_train_lda, X_test_lda = apply_lda(X_train, X_test, y_train, y_test, n_components=9)

# 3. Visualize PCA results
plt.figure(figsize=(12, 5))
for n in n_components_list:
    plt.plot(np.cumsum(pca_results[n]['var_ratio']), 
             label=f'n_components={n}')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('PCA Explained Variance Ratio')
plt.legend()
plt.grid(True)
plt.show()

# 4. Train v evaluate models
def train_and_evaluate(X_train, X_test, y_train, y_test, model_name='SVM'):
    model = SVC(kernel='rbf')
    
    # Training time
    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time
    
    # Prediction time
    start_time = time.time()
    y_pred = model.predict(X_test)
    predict_time = time.time() - start_time
    
    # Accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    return {
        'accuracy': accuracy,
        'train_time': train_time,
        'predict_time': predict_time,
        'classification_report': classification_report(y_test, y_pred)
    }

# Dictionary  lu kt qu
results = {}

# Train trn d liu gc
model_name = 'SVM'
results[f'Original_{model_name}'] = train_and_evaluate(
    X_train, X_test, y_train, y_test, model_name
)

# Train trn d liu PCA
for n in n_components_list:
    results[f'PCA_{n}_{model_name}'] = train_and_evaluate(
        pca_results[n]['X_train'], pca_results[n]['X_test'],
        y_train, y_test, model_name
    )

# Train trn d liu LDA
results[f'LDA_{model_name}'] = train_and_evaluate(
    X_train_lda, X_test_lda, y_train, y_test, model_name
)

# 5. Visualize results
def plot_comparison(results):
    # Prepare data for plotting
    methods = list(results.keys())
    accuracies = [results[m]['accuracy'] for m in methods]
    train_times = [results[m]['train_time'] for m in methods]
    
    # Plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Accuracy comparison
    ax1.bar(range(len(methods)), accuracies)
    ax1.set_xticks(range(len(methods)))
    ax1.set_xticklabels(methods, rotation=45, ha='right')
    ax1.set_ylabel('Accuracy')
    ax1.set_title('Model Accuracy Comparison')
    
    # Training time comparison
    ax2.bar(range(len(methods)), train_times)
    ax2.set_xticks(range(len(methods)))
    ax2.set_xticklabels(methods, rotation=45, ha='right')
    ax2.set_ylabel('Training Time (seconds)')
    ax2.set_title('Training Time Comparison')
    
    plt.tight_layout()
    plt.show()

plot_comparison(results)

# 6. Print detailed results
print("\nDetailed Results:")
for method in results:
    print(f"\n{method}:")
    print(f"Accuracy: {results[method]['accuracy']:.4f}")
    print(f"Training Time: {results[method]['train_time']:.2f} seconds")
    print(f"Prediction Time: {results[method]['predict_time']:.2f} seconds")
    print("\nClassification Report:")
    print(results[method]['classification_report'])
```

```python
from sklearn.decomposition import PCA

# Gi nh rng bn  c hm `reconstruction`  ti to nh t vector PCA

def n_sample_reconstructions(X, n_samples=5, trans_n=120, trans=None):
    sample_indices = np.round(np.random.random(n_samples) * len(X)).astype(int)
    return (sample_indices, 
            np.vstack([reconstruction(X[ind], trans_n, trans) for ind in sample_indices]))

def plot_reconstructions(X, n_samples=5, trans_list=[50, 120, 200], trans=None):
    """
    Plots `n_samples` reconstructions for each specified number of components in `trans_list`.
    """
    fig, axarr = plt.subplots(n_samples, 4, figsize=(16, n_samples * 4))
    
    for i in range(n_samples):
        index = int(np.random.randint(0, len(X)))
        
        # V hnh nh gc
        sns.heatmap(X[index].reshape(28, 28), cmap='gray_r', ax=axarr[i][0])
        axarr[i][0].set_title("Original Image", fontsize=12)
        axarr[i][0].axis('off')
        
        # V cc nh ti to t s lng thnh phn PCA khc nhau
        for j, n_components in enumerate(trans_list):
            recon_img = reconstruction(X[index], n_components, trans).reshape(28, 28)
            sns.heatmap(recon_img, cmap='gray_r', ax=axarr[i][j + 1])
            axarr[i][j + 1].set_title(f"{n_components}-Vector Reconstruction", fontsize=12)
            axarr[i][j + 1].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# Thc hin PCA vi s lng thnh phn ti a cn thit
max_components = 200
pca = PCA(n_components=max_components)
X_pca = pca.fit_transform(X_test)  # Bin i d liu bng PCA

# nh ngha hm reconstruction
def reconstruction(x, n_components, pca):
    """
    Ti to mt hnh nh t mt vector PCA vi s lng thnh phn `n_components`.
    """
    # Chn n_components u tin ca PCA cho vic ti to
    x_reduced = pca.transform([x])[0][:n_components]
    # t cc thnh phn cn li bng 0
    x_full = np.zeros(pca.n_components_)
    x_full[:n_components] = x_reduced
    # Inverse transform  ti to li hnh nh gc
    return pca.inverse_transform([x_full])[0]

# Gi hm plot_reconstructions vi cc s lng thnh phn PCA khc nhau
plot_reconstructions(X_test, n_samples=5, trans_list=[50, 120, 200], trans=pca)
```

```python
import cupy as xp 
import sklearn.model_selection
from sklearn.datasets import load_digits
from svm import SVM

# Load the digits dataset, made up of 1797 8x8 images of hand-written digits
digits = load_digits()

# Divide the data into train, test sets 
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(digits.data, digits.target)

# Move data to GPU
x_train = xp.asarray(x_train)
x_test =  xp.asarray(x_test)
y_train = xp.asarray(y_train)
y_test = xp.asarray(y_test)

# initialize a new predictor
svm = SVM(kernel='rbf', kernel_params={'sigma': 15}, classification_strategy='ovr', x=x_train, y=y_train, n_folds=3, use_optimal_lambda=True, display_plots=True)

svm.fit(x_train, y_train)

# predict things
svm.predict(x_test)

# compute misclassification error
misclassification_error = svm.compute_misclassification_error(x_test, y_test)
print('Misclassification error, lambda = {} : {}\n'.format(svm._lambduh, misclassification_error))

# ******** Output ********
# Misclassification error, lambda = .1 : 0.015555555555555545
```
