# Data loading

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import time

from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
def load_mnist(path, kind='train'):
    import os
    import gzip
    import numpy as np

    """Load MNIST data from `path`"""
    labels_path = os.path.join(path,
                               '%s-labels-idx1-ubyte.gz'
                               % kind)
    images_path = os.path.join(path,
                               '%s-images-idx3-ubyte.gz'
                               % kind)

    with gzip.open(labels_path, 'rb') as lbpath:
        labels = np.frombuffer(lbpath.read(), dtype=np.uint8,
                               offset=8)

    with gzip.open(images_path, 'rb') as imgpath:
        images = np.frombuffer(imgpath.read(), dtype=np.uint8,
                               offset=16).reshape(len(labels), 784)

    return images, labels
```

```python

X_train, y_train = load_mnist('/content/data', kind='train')
X_test, y_test = load_mnist('/content/data', kind='t10k')
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# Classes of Fashion MNIST
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Plot some sample images (need to reshape each sample back to 28x28 for visualization)
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i].reshape(28, 28), cmap='gray')  # Reshape for plotting
    plt.title(class_names[y_train[i]])
plt.show()
```

# Preprocessing data

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
Xmean = X_train.mean(axis=0)
X_train_normalized = X_train - Xmean
X_test_normalized = X_test - Xmean
```

# Training and Evaluation

v Linear kernel khng h tr vi d liu t cupy  train trn GPU nn ta dng d liu bth  train

```python
time_start = time.time()
lin_clf = LinearSVC(random_state=42)
lin_clf.fit(X_train_normalized, y_train)
time_end = time.time()
time_lin_svc = time_end - time_start

print("Training time of Linear SVM: ", time_lin_svc, "seconds")
```

```python
y_pred_linearSVM = lin_clf.predict(X_test_normalized)

train_accuracy_lin_svm = lin_clf.score(X_train_normalized, y_train)
print("Train accuracy of Linear SVM: ", train_accuracy_lin_svm)
```

```python
test_accuracy_lin_svm = lin_clf.score(X_test_normalized, y_test)
print("Train accuracy of Linear SVM: ", test_accuracy_lin_svm)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred_linearSVM)

# Normalize Confusion Matrix
conf_matrix_normalized = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]

plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_normalized, annot=True, fmt='.2f', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Normalized Confusion Matrix')
plt.show()
```

Khi train trn SVM linear th tn kh nhiu thi gian  train nn ta s quyt nh train 10k nh  rt ngn thi gian training, nhng vn cn m bo d liu c nhn balance khi ly sample, hng tip cn ny l **two-stage training**

```python
num_class = 10
sample_per_class = 1000

X_balance = []
y_balance = []

for i in range(num_class):
    class_indices = np.where(y_train == i)[0]
    chosen_indices = np.random.choice(class_indices, sample_per_class, replace=False)

    X_balance.append(X_train_normalized[chosen_indices])
    y_balance.append(y_train[chosen_indices])

X_small = np.vstack(X_balance)
y_small = np.hstack(y_balance)
```

```python
time_start = time.time()
svm_clf = SVC(kernel="rbf", gamma="scale")
svm_clf.fit(X_small, y_small)
time_end = time.time()
print("Training time of SVM with RBF kernel: ", time_end - time_start, "seconds")
```

```python

print("Accuracy on small training set: ", svm_clf.score(X_small, y_small))
```

Sau khi train th vi b d liu nh hn 6 ln b d liu ban u th kt qu cho ra cng  cao hn 0.87365 ca SVM linear khi c train trn tp d liu gc

Chng ta tip tc tinh chnh siu tham s cho model SVM c th  y l parameter C  tm c gi tr C ti u cho model

```python
import warnings
warnings.filterwarnings("ignore")

X_small_df = pd.DataFrame(X_small)
y_small_df = pd.DataFrame(y_small)

Cs = np.logspace(-2, 3, num=10)
MCV_list = []

for C in Cs:
    current_svm = SVC(kernel='rbf', C=C)
    kf = KFold(n_splits=3, random_state=1, shuffle=True)

    mean_crossval = np.mean(cross_val_score(current_svm, X_small_df, y_small_df, cv=kf))
    MCV_list.append(mean_crossval)
    print("On C=", C, "\tMCV=", mean_crossval)

```

```python

sns.set_style('darkgrid')
fig = plt.figure(figsize=(15,10))
lw = 2.5

plt.plot(Cs, MCV_list, color= 'teal', linewidth = lw)
plt.title("Mean Cross Validation accuracy across Penalties" ,fontsize=15, fontweight="bold")
plt.xlabel('Penalty C', fontsize = 15)
plt.xscale("log")
plt.ylabel('Mean Cross Validation', fontsize = 15)
plt.show()
```

Sau  ta tin hnh train trn ton b tp d liu sau khi tm ra c tham s C ti u cho tp d liu nh

chn C=1.6681005372000592 ch khng chn C=5.994842503189409, v rt c th ti im ny th Mean cross validation score  gim nn c th gy ra overfitting

```python
time_start = time.time()
optimal_C= 1.6681005372000592
final_svm = SVC(kernel='rbf', C=optimal_C)
final_svm.fit(X_train_normalized, y_train)
time_end = time.time()
time_svc = time_end - time_start

y_pred_SVM = final_svm.predict(X_test_normalized)
train_accuracy_svm = final_svm.score(X_train_normalized, y_train)
test_accuracy_svm = final_svm.score(X_test_normalized, y_test)

print("Train Accuracy: ", train_accuracy_svm)
print("Test Accuracy: ", test_accuracy_svm)
print("Training time of SVM with RBF kernel on full dataset: ", time_svc, "seconds")
```

```python
conf_matrix = confusion_matrix(y_test, y_pred_SVM)

# Normalize Confusion Matrix
conf_matrix_normalized = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]

plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_normalized, annot=True, fmt='.2f', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Normalized Confusion Matrix')
plt.show()
```

# Dimension reduce by using PCA and Evaluate of reduce models

### Curse of Dimention l g?
Curse of dimetion l hin tng sy ra khi s chiu(hay s lng c trng) trong mt khng gian d liu tng cao. Khi s chiu tng ln cao th khng gian d liu cng m rng ra rt nhanh lm cho d liu tr nn ri rc(Data Sparsity) khin chng cch xa nhau lm vic tnh ton khong cch gia cc im tr nn kh khn hn v lm cho m hinh phn loi bng *Hyperplane* ca SVM tr nn kh khn hn. N cng d gy ra vn  Overfitting hn do c nhiu tham s  hc. ng thi cng lm cho m hinhg kh khn hn trong vic tnh ton.

```python
def pca_reduction(X_train, X_test, target_dim):
    if target_dim:
        pca = PCA(n_components=target_dim, random_state=42)
    else:
        pca = PCA(random_state=42)
    pca.fit(X_train)
    pca_train = pca.transform(X_train)
    pca_test = pca.transform(X_test)
    return pca_train, pca_test

def lda_reduction(X_train, y_train, target_dim=100):
  lda = LDA(n_components=target_dim)
  X_train_reduced = lda.fit_transform(X_train, y_train)

  return X_train_reduced
```

```python
X_train_pca, X_test_pca = pca_reduction(X_train_normalized, X_test_normalized, target_dim=100)
X_train_lda = lda_reduction(X_train_normalized, y_train, 9)
X_test_lda = lda_reduction(X_test_normalized, y_test, 9)
```

```python
Linear_SVM_pca = LinearSVC(random_state=42)
RBF_SVM_pca = SVC(kernel="rbf", gamma="scale")
Linear_SVM_lda =  LinearSVC(random_state=42)
RBF_SVM_lda = SVC(kernel="rbf", gamma="scale")

time_start_pca_linear = time.time()
Linear_SVM_pca.fit(X_train_pca, y_train)
time_end_pca_linear = time.time()

time_start_pca_rbf = time.time()
RBF_SVM_pca.fit(X_train_pca, y_train)
time_end_pca_abf = time.time()

time_start_lda_linear = time.time()
Linear_SVM_lda.fit(X_train_lda, y_train)
time_end_lda_linear = time.time()

time_start_lda_rbf = time.time()
RBF_SVM_lda.fit(X_train_lda, y_train)
time_end_lda_rbf = time.time()

time_pca_linear = time_end_pca_linear - time_start_pca_linear
time_pca_rbf = time_end_pca_abf - time_start_pca_rbf
time_lda_linear = time_end_lda_linear - time_start_lda_linear
time_lda_rbf = time_end_lda_rbf - time_start_lda_rbf
```

```python
print("Training time of Linear SVM with PCA: ", time_pca_linear, "seconds")
print("Training time of RBF SVM with PCA: ", time_pca_rbf, "seconds")
print("Training time of Linear SVM with LDA: ", time_lda_linear, "seconds")
print("Training time of RBF SVM with LDA: ", time_lda_rbf, "seconds")
```

```python
y_pred_pca_linear = Linear_SVM_pca.predict(X_test_pca)
y_pred_pca_rbf = RBF_SVM_pca.predict(X_test_pca)
y_pred_lda_linear = Linear_SVM_lda.predict(X_test_lda)
y_pred_lda_rbf = RBF_SVM_lda.predict(X_test_lda)

train_accuracy_pca_linear = Linear_SVM_pca.score(X_train_pca, y_train)
train_accuracy_pca_rbf = RBF_SVM_pca.score(X_train_pca, y_train)
test_accuracy_pca_linear = Linear_SVM_pca.score(X_test_pca, y_test)
test_accuracy_pca_rbf = RBF_SVM_pca.score(X_test_pca, y_test)

train_accuracy_lda_linear = Linear_SVM_lda.score(X_train_lda, y_train)
train_accuracy_lda_rbf = RBF_SVM_lda.score(X_train_lda, y_train)
test_accuracy_lda_linear = Linear_SVM_lda.score(X_test_lda, y_test)
test_accuracy_lda_rbf = RBF_SVM_lda.score(X_test_lda, y_test)
```

```python
print("Train accuracy of Linear SVM with PCA: ", train_accuracy_pca_linear)
print("Test accuracy of Linear SVM with PCA: ", test_accuracy_pca_linear)
print("Train accuracy of RBF SVM with PCA: ", train_accuracy_pca_rbf)
print("Test accuracy of RBF SVM with PCA: ", test_accuracy_pca_rbf)

print("Train accuracy of Linear SVM with LDA: ", train_accuracy_lda_linear)
print("Test accuracy of Linear SVM with LDA: ", test_accuracy_lda_linear)
print("Train accuracy of RBF SVM with LDA: ", train_accuracy_lda_rbf)
print("Test accuracy of RBF SVM with LDA: ", test_accuracy_lda_rbf)
```

```python
results = {
"Model" : ["Linear SVM", "RBF SVM", "PCA Linear SVM", "PCA RBF SVM", "LDA Linear SVM", "LDA RBF SVM"],
"Train accuracy" : [train_accuracy_lin_svm, train_accuracy_svm, train_accuracy_pca_linear, train_accuracy_pca_rbf, train_accuracy_lda_linear, train_accuracy_lda_rbf],
"Test accuracy" : [test_accuracy_lin_svm, test_accuracy_svm, test_accuracy_pca_linear, test_accuracy_pca_rbf, test_accuracy_lda_linear, test_accuracy_lda_rbf],
"Time traning": [time_lin_svc, time_svc, time_pca_linear, time_pca_rbf, time_lda_linear, time_lda_rbf]
}

df = pd.DataFrame(results)
df.set_index("Model", inplace=True)
print(df)
```

```python
pca = PCA(n_components=2)

Z = pca.fit_transform(X_train_normalized)

V = pca.components_

Z_x = Z[0]
Z_y = Z[1]

std_x = np.std(Z_x)
std_y = np.std(Z_y)

x_min, x_max = -5*std_x, 5*std_x
y_min, y_max = -5*std_y, 5*std_y
x_step = std_x/10
y_step = std_y/10

xx, yy = np.meshgrid(np.arange(x_min, x_max, x_step), np.arange(y_min, y_max, y_step))

z = np.c_[xx.ravel(), yy.ravel()]

# Revert projection
x = np.dot(z,V)

predictions = final_svm.predict(x)
predictions = predictions.reshape(xx.shape)

# Plot decision boundaries
sns.set_style("whitegrid")
plt.figure(figsize = (15,10))
contour = plt.contourf(xx, yy, predictions, cmap=plt.cm.coolwarm)
clb = plt.colorbar(contour, boundaries=list(np.unique(predictions)),
                        ticks=list(np.unique(predictions)),spacing='proportional')
clb.ax.set_title('Class', fontsize=15)
plt.xlabel('1st Principal Component', fontsize = 15)
plt.ylabel('2nd Principal Component', fontsize = 15)
plt.title("Decision surface of Support Vector Machine", fontweight = 'bold', fontsize=15)
```
